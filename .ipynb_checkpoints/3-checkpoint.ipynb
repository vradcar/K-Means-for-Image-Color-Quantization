```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances_argmin
import sys
import os
```

**1. Manually Implement K-Means Algorithm (30 points)**

```python
# K-Means Algorithm
def kmeans(image, k, max_iters=100):
    # Reshape image to 2D array of pixels (as each row is RGB triplet)
    pixels = image.reshape(-1, 3)
    
    # assign centroids randomly by selecting k random pixels
    random_indices = np.random.choice(pixels.shape[0], k, replace=False)
    centroids = pixels[random_indices]
    
    for i in range(max_iters):
        # pixels to the nearest centroid
        labels = pairwise_distances_argmin(pixels, centroids)
        
        # Compute new centroids as the mean of the pixels in each cluster
        new_centroids = np.array([pixels[labels == j].mean(axis=0) for j in range(k)])
        
        # manage empty clusters by reinitializing them randomly
        for j in range(k):
            if np.isnan(new_centroids[j]).any():  # Check for empty clusters
                new_centroids[j] = pixels[np.random.choice(pixels.shape[0])]
        
        # Check if centroids do not change
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids
    
    return centroids, labels
```

**2. Implement Image Color Quantization (20 points)**

```python
# Quantize Image Function
def quantize_image(image, centroids, labels):
    # Replace each pixel with its corresponding centroid's color
    quantized_img = centroids[labels].reshape(image.shape).astype(np.uint8)
    return quantized_img
```

**3. Apply K-Means on test.jpg with K = 2, 8, 16, 32 (20 points)**

```python
# Apply K-Means on test.jpg
def apply_kmeans(image_path, k_values):
    # load image
    image = plt.imread('test.jpg')
    
    plt.figure(figsize=(15, 8))
    
    # Plot original image
    plt.subplot(1, len(k_values) + 1, 1)
    plt.title('Original Image')
    plt.imshow(image)
    plt.axis('off')
    
    # Apply K-Means for each k value
    for i, k in enumerate(k_values):
        centroids, labels = kmeans(image, k)
        quantized_img = quantize_image(image, centroids, labels)
        
        # Plot quantized image
        plt.subplot(1, len(k_values) + 1, i + 2)
        plt.title(f'K = {k}')
        plt.imshow(quantized_img)
        plt.axis('off')
    
    plt.show()
```

**4. Randomly Pick Centroids for Color Quantization (10 points)**

```python
# Random Centroids Quantization
def random_centroids_quantization(image_path, k_values):
    # Load the image
    image = plt.imread('test.jpg')
    pixels = image.reshape(-1, 3)
    
    plt.figure(figsize=(15, 8))
    
    # Plot original image
    plt.subplot(1, len(k_values) + 1, 1)
    plt.title('Original Image')
    plt.imshow(image)
    plt.axis('off')
    
    # Apply random centroid quantization for each k value
    for i, k in enumerate(k_values):
        # Randomly select k centroids from the pixels
        centroids = pixels[np.random.choice(pixels.shape[0], k, replace=False)]
        labels = pairwise_distances_argmin(pixels, centroids)
        
        quantized_img = quantize_image(image, centroids, labels)
        
        # Plot quantized image
        plt.subplot(1, len(k_values) + 1, i + 2)
        plt.title(f'Random K = {k}')
        plt.imshow(quantized_img)
        plt.axis('off')
    
    plt.show()
```

**5. Pick a Custom Image and Conduct Image Color Quantization Using K-Means (5 points)**

```python
# Apply to a Custom Image batman.jpg
def quantize_custom_image(image_path, k):
    # Load a custom image
    image = plt.imread('batman.jpg')
    
    # Apply K-Means quantization
    centroids, labels = kmeans(image, k)
    quantized_img = quantize_image(image, centroids, labels)
    
    # Plot the quantized image
    plt.figure(figsize=(8, 8))
    plt.title(f'K-Means Quantization (K = {k})')
    plt.imshow(quantized_img)
    plt.axis('off')
    plt.show()
```

**6. Test the Implementation and Compare Images Generated Using K-Means and Random Centroids (5 points)**

```python
# Test the implementation:
image_path = 'test.jpg' 

# K values for quantization
k_values = [2, 8, 16, 32]

# Apply K-Means quantization
apply_kmeans(image_path, k_values)

# Apply random centroid quantization
random_centroids_quantization(image_path, k_values)

# Apply to custom image 
custom_image_path = 'batman.jpg' 
quantize_custom_image(custom_image_path, k=8)
```
